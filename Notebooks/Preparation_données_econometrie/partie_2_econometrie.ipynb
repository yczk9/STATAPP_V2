# Construction des jeux de données pour le Staggered DiD

Import des bibliothèques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import des données

```python
df_caractéristiques = pd.read_csv(r"C:\Users\yancr\Documents\ENSAE_V2\STATAPP\V2\STATAPP_V2\Données\données_caracteristiques_communes.csv", index_col=0)
df_caractéristiques['codecommune'] = df_caractéristiques['codecommune'].astype(str).str.zfill(5)

df_sirene_social = pd.read_csv(r"C:\Users\yancr\Documents\ENSAE_V2\STATAPP\V2\STATAPP_V2\Données\Sirene_social.csv", index_col=0)
df_sirene_social['codeCommuneEtablissement'] = df_sirene_social['codeCommuneEtablissement'].astype(str).str.zfill(5)

df_sirene_transactionnel = pd.read_csv(r"C:\Users\yancr\Documents\ENSAE_V2\STATAPP\V2\STATAPP_V2\Données\Sirene_transactionnel.csv", index_col=0)
df_sirene_transactionnel['codeCommuneEtablissement'] = df_sirene_transactionnel['codeCommuneEtablissement'].astype(str).str.zfill(5)
```

```python
df_sirene_social = df_sirene_social.rename(columns={'codeCommuneEtablissement' : 'codecommune'})
df_sirene_transactionnel = df_sirene_transactionnel.rename(columns={'codeCommuneEtablissement' : 'codecommune'})                     
```

```python
Liste_equipements_social = ['Hôtels et hébergement', 'Campings',
       'Restauration traditionnelle', 'Débits de boissons', 'Librairie',
       'Presse et papeterie']

Liste_equipements_transactionnel = ['Restauration rapide', 'Alimentation générale',
       'Supérettes', 'Autres commerces non spécialisés', 'Vente à distance',
       'Banques et établissements financiers', 'Pharmacies']
```

## Rural non périurbain

### Etablissements 'social'

On selectionne uniquement les communes classées rurale non périurbaine

```python
df_communes_social_rnp = df_caractéristiques.loc[df_caractéristiques['typologie_urbaine'] == 'Rural non périurbain']
```

On merge avec les données concernant la présence d'équipements

```python
df_communes_social_rnp = df_communes_social_rnp.merge(df_sirene_social, how='inner', on = ['codecommune','Annee'])
```

On calcule le total d'équipement par commune et par année

```python
df_communes_social_rnp['total_equipements'] = df_communes_social_rnp[Liste_equipements_social].sum(axis=1)
```

On supprime les colonnes qui ne sont plus nécessaires

```python
df_communes_social_rnp = df_communes_social_rnp.drop(columns=['typologie_urbaine']+Liste_equipements_social)
```

Export du DataFrame

```python
df_communes_social_rnp.to_csv("communes_social_rnp.csv")
```
